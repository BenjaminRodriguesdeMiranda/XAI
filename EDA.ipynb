# EDA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("data/german_credit_data.csv")
df_encoded = pd.read_csv("data/encoded_dataset.csv")
```

```python
df
```

```python
df.nunique()
```

```python
sns.set(rc = {'figure.figsize':(8,6)})
g1 = sns.countplot(x="Risk",data= df)                  
g1.set_title("Target variable(Risk) Distribution", fontsize=15)
g1.set_xlabel("Risk")
g1.set_xlabel("Count")
plt.show()
```

## Checking for correlation

```python
corr = df.corr()

_ = sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns, annot=True)
```

```python
corr = df_encoded.corr()

plt.figure(figsize=(30,30))
_ = sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns, annot=True)
```
